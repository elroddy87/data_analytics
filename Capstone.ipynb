```python
import pandas as pd
import numpy as np 
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats
import matplotlib.pyplot as plt
import datetime as dt
from datetime import datetime, timedelta,date
import warnings
import sqlite3
import dataquality as dq
```

```python
#Perform descriptive statistical analysis for numerical features

#Using dataquality.py file to perform Statistical analyis

diabetic_data = pd.read_csv('diabetic_data.csv')
dq.data_check(diabetic_data)
```

```python
#Visualize the distribution of categorical features race and gender
race_gender_subset = diabetic_data[['race','gender', 'readmitted']]
race_gender_subset_grp = race_gender_subset.groupby(['race', 'gender'])['readmitted'].agg(['count']).reset_index()

print(race_gender_subset_grp)
sns.barplot(data=race_gender_subset_grp, x='race', y='count', hue='gender')

# Add titles and labels
plt.title('Readmission Count by Race and Gender')
plt.xlabel('Race')
plt.ylabel('Number of Readmissions')

plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Explore the relationship between readmission status and age
plt.figure(figsize=(8,5))
sns.countplot(x='age', hue='readmitted', data=diabetic_data)
plt.title('Readmission Count by Age Group')
plt.xticks(rotation=90)
plt.show()
```

```python
#Investigate correlations between numerical features

#Create a subset with numerical values
diabetic_data_numerical = diabetic_data[ ['time_in_hospital',
       'num_lab_procedures', 'num_procedures', 'num_medications',
       'number_outpatient', 'number_emergency', 'number_inpatient',
       'number_diagnoses']]
print(diabetic_data_numerical.columns)
#Create Correlation Matrix for numerical values
corr_matrix = diabetic_data_numerical.corr()
#print(corr_matrix)
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix of Numerical Features")
plt.show()

# Summary: The below variables have a moderate positive correlation relationship:
#    - time_in_hosipital / num_medications > May indicate people on more medications spend more time in the hospital
#   - time_in_hosipital / num_lab_procedures > May indicate people who have more lab procedures spend more time in the hospital
#   - num_medication / num_procedures > May indicate people with more procedures take increase number of medications
#    
```

```python
#Analyze the distribution of medication changes and total medications taken

# Based on the scatter plot/trend line and correlation analysis, distribution of medication changes and total medications taken have a strong
# positive correlation.  Increase number of medications = more medication changes

medication_subset = diabetic_data[['patient_nbr', 'num_medications', 'change']]
medication_subset_grp = medication_subset.groupby('patient_nbr').agg(total_changes=('change', 'count'),total_meds=('num_medications', 'sum')).reset_index()
medication_subset_grp

#Get Correlation
corr = medication_subset_grp['total_changes'].corr(medication_subset_grp['total_meds'])
print(f"Correlation between total_changes and total_meds: {corr:.2f}")

#plot
sns.lmplot(data=medication_subset_grp, x='total_changes', y='total_meds', line_kws={'color': 'red'})
plt.title('Trend: Medication Changes vs Total Medications')
plt.show()

```

```python
#Examine the distribution of diagnoses categories

#Combine diag1, diag2, and diag3 to get total count
diag1_category_counts =  diabetic_data['diag_1'].value_counts().reset_index().rename(columns={'diag_1': 'diagnosis'}) 
diag2_category_counts =  diabetic_data['diag_2'].value_counts().reset_index().rename(columns={'diag_2': 'diagnosis'}) 
diag3_category_counts =  diabetic_data['diag_3'].value_counts().reset_index().rename(columns={'diag_3': 'diagnosis'}) 

concat_data = pd.concat([diag1_category_counts,diag2_category_counts,diag3_category_counts])
all_diagnosis = concat_data.groupby('diagnosis', as_index=False)['count'].sum()

#Show Top 10 Diagnosis Types
top10 = all_diagnosis.nlargest(10, 'count')

plt.figure(figsize=(8,5))
sns.barplot(x='diagnosis', y='count', data=top10)
plt.title('Top 10 Most Frequent Diagnoses')
plt.xlabel('Diagnosis Code')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Explore the distribution of patients across admission types, sources, and discharge dispositions

#Admission Type Distribution
admission_type_subset = diabetic_data[['admission_type_id']]
admission_type_grp = admission_type_subset.groupby('admission_type_id').size().reset_index(name='count')

admission_source_subset = diabetic_data[['admission_source_id']]
admission_source_grp = admission_source_subset.groupby('admission_source_id').size().reset_index(name='count')

discharge_disposition_subset = diabetic_data[['discharge_disposition_id']]
discharge_disposition_grp = discharge_disposition_subset.groupby('discharge_disposition_id').size().reset_index(name='count')

plt.figure(figsize=(8,5))
sns.barplot(x='admission_type_id', y='count', data=admission_type_grp)
plt.title('Admission Type Distribution')
plt.xlabel('Admission Type Id')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(x='admission_source_id', y='count', data=admission_source_grp)
plt.title('Admission Source Distribution')
plt.xlabel('Admission Source Id')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(x='discharge_disposition_id', y='count', data=discharge_disposition_grp)
plt.title('Discharg Disposition Distribution')
plt.xlabel('Discharge Disposition Id')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()



```
